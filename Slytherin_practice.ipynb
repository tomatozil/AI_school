```python
list_name = []
list_house = []
list_description = []

webpage = requests.get('https://namu.wiki/w/슬리데린')
soup = BeautifulSoup(webpage.content, "html.parser")
for td in soup('td') : # td 안의
    for a in td('a') : # a 태그 중에서
        if ('class' in a.attrs # a가 class 속성을 가지고 있고
            and 'wiki-link-internal' in a['class']) : #class 기 wiki-link-internal 이다.
            list_name.append(a['title'])
            list_house.append("슬리데린")

print("진행 중")
for name in list_name:
    webpage = requests.get('https://namu.wiki/w/' + name)
    soup = BeautifulSoup(webpage.content, "html.parser")
    name_contents = soup.find('div', {'class':'w'})
    list_description.append(name_contents.text)
print("완료")
```

```python
# 리스트 → 딕셔너리 (슬리데린 기준)
dict_slytherin = {
    "name" : list_name,
    "house" : list_house,
    "description" : list_description
}
```

```python
word_cleaned_dic = {}

for word in word_cleaned :
    if word not in word_cleaned_dic :
        word_cleaned_dic[word] = 1
    else :
        word_cleaned_dic[word] += 1
word_cleaned_dic
```
