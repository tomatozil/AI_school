```python
# 데이터 처리
import pandas as pd
import numpy as np

# 텍스트 처리
import nltk
from nltk.corpus import stopwords
from collections import Counter

# 시각화
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from PIL import Image
%matplotlib inline
```

```python
df = pd.read_excel("data_harrypotter.xlsx")
```

```python
# 각 기숙사의 wordcloud 그려보기

lines = df[df['house'] == "그리핀도르"]['description'].tolist()

tokens = []

for line in lines:
    tokenized = nltk.word_tokenize(line)
    for token in tokenized :
        if (token != "때문") & (token != "없다") & (token != "이자"):
            tokens.append(token)
        
word_frequency = nltk.FreqDist(tokens)

masking = np.array(Image.open("그리핀도르.jpeg")) 
image_colors = ImageColorGenerator(masking)

word_cloud = WordCloud(font_path="NanumSquareOTF_acB.otf",
                       min_font_size=5, max_font_size=38,
                       width=1600, height=1000,
                       mask=masking,
                       background_color="black").generate_from_frequencies(word_frequency)

plt.figure(figsize=(10,10))  
plt.imshow(word_cloud.recolor(colormap = "hot_r"), interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
